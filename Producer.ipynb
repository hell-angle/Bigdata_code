```python
from kafka import KafkaProducer
import csv
import time
import json
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import pickle
```

```python
# Step 1: Load the dataset
data = pd.read_csv('D:\\Big_data\\Project_Bigdata\\streaming.csv')

# Select only the desired features
selected_features = ['step', 'amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
data_selected = data[selected_features]

# Kafka configuration
bootstrap_servers = '192.168.1.27:9092'
topic = 'test'
```

```python

data = data.drop(['type', "nameOrig", "nameDest", "isFraud", "isFlaggedFraud"], axis=1)
load_model1 = pickle.load(open("Random_forest.sav", "rb"))
load_model2 = pickle.load(open("Decision_tree.sav", "rb"))
load_model3 = pickle.load(open("Logistic_regression_model.sav", "rb"))

# Create a Kafka producer
producer = KafkaProducer(bootstrap_servers=bootstrap_servers, value_serializer=lambda v: v.encode('utf-8'))
```

```python
for _, row in data_selected.iterrows():
    # Perform prediction using the loaded models
    prediction_rf = load_model1.predict([row.values])[0]
    prediction_dt = load_model2.predict([row.values])[0]
    prediction_lr = load_model3.predict([row.values])[0]
    #prediction_dt = decision_tree.predict([row.values])[0]
    #prediction_lr = logistic_regression_model.predict([row.values])[0]

    # Perform actions based on the predictions
    if prediction_rf == 1:
        message = 'Random Forest model found fraud'
    elif prediction_dt == 1:
        message = 'Decision Tree model found fraud'
    elif prediction_lr == 1:
        message = 'Logistic Regression model found fraud'
    else:
        message = 'All models not found fraud'
    # Send the message to the Kafka topic
    producer.send(topic, value=message)
    time.sleep(0.01)

# Close the Kafka producer
producer.close()
```
